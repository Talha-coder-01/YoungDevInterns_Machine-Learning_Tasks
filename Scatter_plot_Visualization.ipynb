```python
# Import necessary libraries for plotting and data manipulation
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the 'tips' dataset from seaborn
tips = sns.load_dataset('tips')
```

```python
# Choose 'total_bill' and 'tip' as the features to plot
x = tips['total_bill']
y = tips['tip']
```

```python
# Create a scatter plot with 'total_bill' on the x-axis and 'tip' on the y-axis
plt.scatter(x, y)
plt.xlabel('Total Bill ($)'); plt.ylabel('Tip ($)'); plt.title('Scatter Plot of Total Bill vs Tip'); plt.show()
```

```python
# Calculate the Pearson correlation coefficient between 'total_bill' and 'tip'
print(f'Pearson correlation coefficient: {tips["total_bill"].corr(tips["tip"]):.2f}')
```
